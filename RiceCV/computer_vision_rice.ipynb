```python
import cv2
import os
import numpy as np

# Load the training images
def load_test_images(folder_name):
    images = []
    for file_name in os.listdir(folder_name):
        file_path = os.path.join(folder_name, file_name)
        image = cv2.imread(file_path)
        images.append(image)
    return images

train_images = load_training_images("test")
train_images
```

```python
# Pre-process the training images to extract features
def pre_process_image(image):
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    ret, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)
    contours, _ = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    return len(contours)

processed_train_images = [pre_process_image(image) for image in train_images]
processed_train_images
```

```python
# Print the number of grains in each image
for i, count in enumerate(processed_train_images):
    print(f"Image {i+1}: {count} grains")
```

```python
import cv2
import numpy as np
import os

# Path to the train folder
train_folder = "train"

# Dictionary to store the count of total and broken rice grains for each image
counts = {}

# Loop over all images in the train folder
for image_file in os.listdir(train_folder):
    # Read the image
    image = cv2.imread(os.path.join(train_folder, image_file))

    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Threshold the image to create a binary image
    ret, thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)

    # Find contours in the binary image
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Initialize the count of total and broken rice grains
    total_count = 0
    broken_count = 0

    # Loop over all contours
    for contour in contours:
        # Check if the contour is a broken rice grain
        if len(contour) < 5:
            broken_count += 1
        else:
            total_count += 1

    # Store the count of total and broken rice grains for the image
    counts[image_file] = (total_count, broken_count)

# Print the counts
print(counts)
```
